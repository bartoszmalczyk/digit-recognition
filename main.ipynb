```python
import tensorflow as tf
from keras.datasets import mnist
import numpy as np  
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train_reshaped = x_train.reshape(x_train.shape[0], -1)
x_test_reshaped = x_test.reshape(x_test.shape[0], -1)

x_train_reshaped = x_train_reshaped.astype("float32") / 255
x_test_reshaped = x_test_reshaped.astype("float32") / 255
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])
```

```python
model.compile(loss = "sparse_categorical_crossentropy", metrics=["accuracy"]) 
```

```python
history = model.fit(x_train_reshaped, y_train, epochs = 10)
```

```python
test_loss, test_acc = model.evaluate(x_test_reshaped, y_test)


print(f"\nAccuracy: {test_acc * 100:.2f}%")
```

```python
model.save('digit_recognition.keras')
```
